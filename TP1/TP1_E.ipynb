### La version python de `dicho`

```python
def dicho(a,b,f,n):
    m=(a+b)/2
    err=abs(b-a)
    while err>n :
        if f(m)==0:
            return m
        if f(a)*f(m)<0:
            b=m
        else:
            a=m
    return m
```

## Version tant que qui renvoi la solution avec le nombre d'iteration

```python
def dichow (a ,b,f,eps):

    g,d = min (a,b),max(a,b)
    ga,dr = f(g),f(d)
    n = 0
    while d-g > eps :
        n += 1
        m = ( g + d )/2
        valm = f ( m )
        if ga * valm < 0 :
            d = m
            dr = valm
        else :
            g = m
            ga = valm
    
    return (g+d)/2,n
```

#### 1) <h4>

```python
def newtonn(f,fprime,a,eps):
    x1=a
    x2=x1-f(x1)/fprime(x1)
    while abs(x1-x2)>eps :
        x1=x2
        x2=x1-f(x1)/fprime(x1)
    return x2
```

```python
import math
from sympy import *
from sympy.abc import x

f=lambda x: math.exp(-x)-x
fprime=lambda x: -math.exp(-x)-1
newtonn(f,fprime,0,0.001)
```
